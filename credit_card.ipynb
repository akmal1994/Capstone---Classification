```python
neg_cols = ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6', 'PAY_0','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df[neg_cols] = df[neg_cols].abs()
print(round((df['BILL_AMT1'] < 0).sum()/len(df)*100, 2))
print(round((df['BILL_AMT2'] < 0).sum()/len(df)*100, 2))
print(round((df['BILL_AMT3'] < 0).sum()/len(df)*100, 2))
print(round((df['PAY_0'] < 0).sum()/len(df)*100, 2))
print(round((df['PAY_2'] < 0).sum()/len(df)*100, 2))
print(round((df['PAY_3'] < 0).sum()/len(df)*100, 2))
```

```python
cols = ['SEX', 'MARRIAGE', 'EDUCATION', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
df[cols]=df[cols].astype(object)
df.info()
```

```python
df = pd.get_dummies(df, columns=cols, drop_first=True)
df.shape
```

```python
df.head()
```

```python
X = df.drop(['defaulted'], axis=1)
y = df.defaulted
X.head()
```

```python
from imblearn.combine import SMOTETomek
smt = SMOTETomek(sampling_strategy=0.4, random_state=42)
X_res, y_res = smt.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res, test_size=0.3, random_state=12)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
rf = RandomForestClassifier()
dt = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': range(5, 15, 10),
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf': range(50, 150, 50),
    'min_samples_split': range(50, 150, 50),
}

dtree = DecisionTreeClassifier()

grid_search = GridSearchCV(estimator = dtree, 
                           param_grid = param_grid, 
                           scoring= 'recall',
                           cv = 5, 
                           verbose = 1)
grid_search.fit(X_train,y_train)

dt_model = grid_search.best_estimator_
```

```python
param_grid = param_grid = { 
    'n_estimators': [200, 300,400,500,600],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}


RF = RandomForestClassifier(random_state = 100)


gridcv_rf = GridSearchCV(estimator=RF, param_grid=param_grid, 
                    cv=3,verbose=True)
gridcv_rf.fit(X_train, y_train)
rf_model = gridcv_rf.best_estimator_
rf_model

```

```python
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)  
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(classification_report(y_test, dt_pred))
```

```python
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
print(classification_report(y_test, rf_pred))
```
